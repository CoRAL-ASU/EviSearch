# Baseline results: Gemini File Search

Load evaluation results from `experiment-scripts/baselines_gemini_file_search/` and report:

- **By PDF**: overall correctness, completeness, and overall score per document
- **By category**: exact_match, numeric_tolerance, structured_text — aggregated and per-PDF

```python
import json
from pathlib import Path
from typing import Optional
import pandas as pd

# Base path: run from repo root, experiment-analysis/, or experiment-scripts/
_candidates = [
    Path.cwd() / "experiment-scripts" / "baselines_gemini_file_search",
    (Path.cwd() / ".." / "experiment-scripts" / "baselines_gemini_file_search").resolve(),
    Path.cwd() / "baselines_gemini_file_search",
]
BASELINE_BASE = next((p for p in _candidates if p.exists()), None)
if BASELINE_BASE is None:
    raise FileNotFoundError("Could not find baselines_gemini_file_search. Run from repo root or experiment-analysis/.")

def load_summary_metrics(trial_dir: Path) -> Optional[dict]:
    path = trial_dir / "evaluation" / "summary_metrics.json"
    if not path.exists():
        return None
    return json.loads(path.read_text(encoding="utf-8"))

# Discover all trial directories
trial_dirs = sorted([d for d in BASELINE_BASE.iterdir() if d.is_dir()])
print(f"Found {len(trial_dirs)} trials in {BASELINE_BASE.name}")
```

```python
# Load summary metrics for all trials
rows_by_pdf = []
rows_by_category = []

for trial_dir in trial_dirs:
    data = load_summary_metrics(trial_dir)
    if data is None:
        continue
    pdf_name = trial_dir.name
    overall = data.get("overall", {})
    rows_by_pdf.append({
        "pdf": pdf_name,
        "avg_correctness": overall.get("avg_correctness"),
        "avg_completeness": overall.get("avg_completeness"),
        "avg_overall": overall.get("avg_overall"),
        "total_columns": overall.get("total_columns"),
    })
    for cat_name, cat_metrics in data.get("by_category", {}).items():
        rows_by_category.append({
            "pdf": pdf_name,
            "category": cat_name,
            "avg_correctness": cat_metrics.get("avg_correctness"),
            "avg_completeness": cat_metrics.get("avg_completeness"),
            "avg_overall": cat_metrics.get("avg_overall"),
            "column_count": cat_metrics.get("column_count"),
        })

df_by_pdf = pd.DataFrame(rows_by_pdf)
df_by_category = pd.DataFrame(rows_by_category)
```

## By PDF: overall correctness, completeness, overall score

```python
# Format as percentages for readability
df_display = df_by_pdf.copy()
for col in ["avg_correctness", "avg_completeness", "avg_overall"]:
    df_display[col] = (df_display[col] * 100).round(1).astype(str) + "%"
df_display
```

## By category: exact_match, numeric_tolerance, structured_text (aggregated)

```python
agg = df_by_category.groupby("category").agg(
    avg_correctness=("avg_correctness", "mean"),
    avg_completeness=("avg_completeness", "mean"),
    avg_overall=("avg_overall", "mean"),
    total_column_count=("column_count", "sum"),
).reset_index()
agg_display = agg.copy()
for col in ["avg_correctness", "avg_completeness", "avg_overall"]:
    agg_display[col] = (agg_display[col] * 100).round(1).astype(str) + "%"
agg_display
```

## By category per PDF (overall score pivot)

```python
pivot_overall = df_by_category.pivot(index="pdf", columns="category", values="avg_overall")
(pivot_overall * 100).round(1)
```

## Correctness and completeness by category per PDF

```python
pivot_correctness = df_by_category.pivot(index="pdf", columns="category", values="avg_correctness")
pivot_completeness = df_by_category.pivot(index="pdf", columns="category", values="avg_completeness")
print("Avg correctness by PDF × category (0–1):")
display((pivot_correctness * 100).round(1))
print("Avg completeness by PDF × category (0–1):")
display((pivot_completeness * 100).round(1))
```
